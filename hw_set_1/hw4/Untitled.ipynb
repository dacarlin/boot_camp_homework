```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import pandas 
plt.style.use( 'ggplot' )

df = pandas.read_csv( 'fib_out.txt', sep='\t' )

plt.scatter( df.fib_number, df.run_time, color='red', lw=0 )
plt.title( 'fib_recursive (average of $10^2$ iterations)')
plt.xlabel( 'Fibonacci number' )
plt.ylabel( 'Runing time (second)' )
```
