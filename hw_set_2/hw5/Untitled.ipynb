```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import pandas 
plt.style.use( 'ggplot' )

df = pandas.read_csv( 'fib_out.txt', sep='\s+' )

plt.scatter( df.fib_number, df.run_time * 1e6, color='red', lw=0 )
plt.title( 'fig_iterative (average of $10^7$ iterations) ')
plt.xlabel( 'Fibonacci number' )
plt.ylabel( 'Runing time (microsecond)' )
```
